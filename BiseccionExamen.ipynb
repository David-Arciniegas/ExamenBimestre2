```python
import scipy.optimize as opt
import matplotlib.pyplot as plt
import numpy as np


def funcion(x):

    return x**5 - 6*(x**4) + 2*(x**3) + 20*(x**2) - 27*x + 10

a = -3
b = 6




raiz = opt.bisect(funcion, a, b)

print(f"La raíz en el intervalo [{a}, {b}] es: {raiz}")

x = np.linspace(a, b, 100)


plt.plot(x, funcion(x))
plt.grid()

```
